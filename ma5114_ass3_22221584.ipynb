#### Question 1:


For this question I decided to use a function defined below to track the sequence provided in the question. Each time a match was found a counter ticked up and printed the presence of a binding site, as well as the matching sequence and the start/stop location of said sequences.

```python
import re
def find_CTCF_motif(sequence):
    motif_regex = re.compile(r'CC...AG..GG') #The sequence we're searching for
    runs = re.finditer(motif_regex,sequence) #finditer used because we want to scan entire code
    match_count=0 #used to keep track of the number of hits
    
    for match in runs:
        match_count += 1
        run_start = match.start()
        run_end = match.end()
        match_seq = sequence[run_start:run_end] #isolates the part of the string corresponding to the ctcf motif for later printing
        print("CTCF Motif from " + str(run_start) + " to " +str(run_end))
        print("Seq is: "+match_seq)
        print("")
    print(f"{match_count} CTCF motifs detected in file") 





with open('/home/josh/Documents/MA5114_ProgrammingforBio/intron.txt') as in_file:
    for line in in_file:
        find_CTCF_motif(line) #reads in the file and iterates it through the above function
        
    
    
    
    
```

#### Question 2.

Here I created a small extract_exon function which takes two variables, a list and a string, and uses them to return a list of sequences which can then be cocatenated later. I believe it would have been easier to create one large function that accepted files as inputs, but I don't think the question was asking for this.

```python
def extract_exon(this_exon,sequence):
    
    start = int(this_exon[0])
    end = int(this_exon[1])
    match_seq = sequence[start:end]
    return match_seq








exon_list = []

with open('/home/josh/Documents/MA5114_ProgrammingforBio/exons.txt') as exon_file:
    for line in exon_file:
        line = line.strip()
        line = line.split(',')
        exon_list.append(line) #creates a list of list corresponding to the read-in exons

genomicRNA = ""
with open('/home/josh/Documents/MA5114_ProgrammingforBio/genomic_RNA.fa') as rna_file:
    for line in rna_file:
        if line != "" and line[0].isalpha():
            genomicRNA = line + genomicRNA #appends every line in the genomic_RNA.fa file that starts as a letter to a string
        

splicedRNA=""            
for exon in exon_list: #runs through the list of exons, chopping out matching sequence from the genomicRNA
    rna = extract_exon(exon,genomicRNA)
    splicedRNA = rna + splicedRNA #appends each match to a new string


with open('genomic_mRNA.fa','w') as out_file:
    print(splicedRNA, file = out_file) #prints out the resulting spliced sequence to a .fa file
```
